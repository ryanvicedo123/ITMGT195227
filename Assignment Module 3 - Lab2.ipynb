## 1A. Rectangle

```python


# GRADED FUNCTION

def print_rectangle(width, height):

    '''

    (int, int) -> None

    

    This function prints a solid rectangle composed of asterisks ("*").

    This rectangle should be of the specified width and the specified height.

    This function does not return any value; it simply prints the rectangle string.

    '''

    

    rectangle_string = ""

    

    # Write your code below
    i=0
    ast="*"
    nextline="\n"
    while i<height:
        j=0
        for j in range(0,width):
            rectangle_string=rectangle_string+ast
            j+=1
            if j==width:
                rectangle_string=rectangle_string+nextline
                i+=1
            else:
                continue

    # End of code

    

    print(rectangle_string)

print_rectangle(10,10)
```

## 1B. Hollow Rectangle

## 2. Caesar Cipher

```python
# GRADED FUNCTION

def caesar_encrypt(message, shift):

    '''

    (str, int) -> str

    

    This function encrypts a message using the Caesar cipher method.

    The Caesar cipher "shifts" each letter in a message to the right by a specified number.

    

    For example:

    

    encrypt('ABC', 2) -> 'CDE'

    

    In terms of ordinal positions in the alphabet, 'A' is 0, 'B' is 1, and 'C' is 2.

    Add the specified number to each of these values to receive 0 + 2 = 2, 2 + 2 = 4, and 3 + 2 = 5.

    Converting back to letters from numbers: 2 is 'C', 3 is 'D', and 4 is 'E'.

    

    Another example:

    

    encrypt('XYZ', 2) -> 'ZAB'

    

    In terms of ordinal positions in the alphabet, 'X' is 23, 'Y' is 24, and 'Z' is 25.

    Add the specified number to each to receive 25, 26, and 27 respectively.

    26 and 27 no longer fall in the range of alphabet numbers, so we need to "wrap" them around.

      Subtract 26 from 26 and 27 to receive 0 and 1.

    The final list of numbers is 25, 0, and 1.

    Converting back to letters from numbers: 25 is 'Z', 0 is 'A', and 1 is 'B'.

    

    You may find these functions helpful:

    chr() converts a Unicode integer value to a character (i.e. a letter).

      e.g. chr(65) -> 'A'

    ord() converts a character (i.e. a letter) to its Unicode integer value.

      e.g. ord('A') -> 65

      

    For your convenience:

    1. The message is composed entirely of capital letters. There will be no spaces or lower case letters.

    2. The shift number will be between 0 and 25.

    3. The alphabet is mapped such that A is 0 and Z is 25.

    '''

    

    output = ''

    

    # Write your code below
    
    # Importance of functionality is in creaing a formula 
    # adding the shift to the old UNICODE to create the new UNICODE...
    
    # Given A = 65 in UNICODE, 65 = base negator 
    # Z = 25, A = 26. Given this, 26 = base divider
    
    # 1 - Know and get the corresponding UNICODE for a character (A = 65, B = 66,...)
    # 2 - Subtract the UNICODE value by 65 (the base negator) to get the digit place order
    # 3 - Add the key shift to the digit place order number
    # 4 - Get the modulo (MOD) of #3 to get the remainder, which should be the value of the shift!!
    # 5 - Add 65 to the key shift remaineder to convert the value back to UNICODE.
    # 6 - Use new UNICODE value in conversion functions!!
    
    for letter in str(message):
        output = output + chr((ord(letter) + shift - 65) % 26 + 65)
    
    # Simple, but beautiful....
    
    # End of code

    

    return output

caesar_encrypt("RYANISHANDSOME",6)
```

## 3. Vigenere Cipher

```python
# GRADED FUNCTION

def vigenere_encrypt(message, keyword):

    '''

    (str, str) -> str

    

    This function encrypts a message using the Vigenere cipher method.

    

    The Vigenere cipher is an extension of the Caesar cipher method.

    Instead of using a uniform number, the Vigenere cipher uses a keyword to shift letters.

    

    Example:

    

    vigenere_encrypt('ABC', 'KEY') -> 'KFA'

    

    First, convert the message and the key to numbers.

    'ABC' -> [0, 1, 2]

    'KEY' -> [10, 4, 24]

    Second, add the respective numbers to each other.

    [0, 1, 2] + [10, 4, 24] = [10, 5, 26]

    Third, correct the numbers that do not fall within 0-25 by subtracting 26.

    [10, 5, 26] -> [10, 5, 0]

    Fourth, convert the numbers back to letters.

    [10, 5, 0] -> 'KFA'

    

    For your convenience:

    1. You are highly encouraged to use your Caesar cipher function from Item 2 in this item.

    2. You may assume that the key is the same length as the message.

    

    Bonus points:

    1. +2 (with a perfect base function) (all or nothing) if you adjust for the case that the key is 

      shorter than the message. (e.g. "MEAMORE", "UEDA" -> "MEAMORE", "UEDAUED")

    '''

    

    output = ''

    

    # Write your code below
    
    # TWO DIFFERENT PROCESS PATHS:
    # (1) DECIMALS > ADDITION > MOD26 > DEICMAL to UNICODE (+65 or +97) > STRING
    # (2) UNICODE > ADDITION > MOD26 > DECIMAL to UNICODE (+65 or +97) > STRING
    
    # Whether or not it's decimal or UNICODE, utilizing MOD26 will get the decimal value of the wrap-around!

    # CALCULATION STEPS:
    # 1. Add UNICODE/DECIMAL numbers of keyword and message together
    # 2. Get MOD26 of list to get DECIMAL values of new letters
    # 3. Add 65 to convert DECIMAL values of new letters back to UNICODE
    
    # Functionality is similar to Caesar Cipher, however the decimal/UNICODE values of the message and 
    # keyword must be added to each other BEFORE dividing by the MOD26... 
    
    # For the extra challenge, 'keyword' needs to be functional as well with 'message' in a for-loop 
    # despite their difference in character length. Therefore:
    
    # 1) An iteration system must be used that accounts for BOTH string inputs
    # 2) An array for each of the letters in 'keyword' can alternate between the different characters. The 
    #    characters are conveniently in order and the keyword wraps around in its letter usage.
    # 3) The length of the keyword must be a condition for calling values in the array of the keyword
    
# ===================================================================================================================

#              !!!     letter_decimal = (plaintext_int[i] + key_as_int[i % key_length]) % 26   !!!


# 1. plaintext_int[i]              is each character in UNICODE from message is taken in order placed in the array
#
# 2. key_as_int[i % key_length]    is each character in UNICODE from keyword is taken in order placed in the array
#                                  and KEPT WITHIN BOUNDARIES of the length of the keyword using MOD(keylength).

# 3. % 26                          the whole line is  put under MOD26 to get remainder value, a decimal 

# ===================================================================================================================
    
    key_length = len(keyword)                   # Using the length of the keyword gets us the base divider for MOD
    key_unicode = [ord(i) for i in keyword]     # An array of UNICODE is made based on each of the letters in keyword
    message_unicode = [ord(i) for i in message] # An array of UNICDODE is made based on each the letters in the message
    
    # We can combine the UNICODE values of each letter in a for-loop by expressing them as elements in lists:
    
    # Loop only works based on the length of the message
    for i in range(len(message)):
        
        # THE GENIUS LINE!
        decimal_letters = (message_unicode[i] + key_unicode[i % key_length]) % 26  
        
        # We add 65 to convert back to UNICODE, then output using chr() to get a string character
        output += chr(decimal_letters + 65)  
    
    # End of code

    

    return output
vigenere_encrypt("RYANISHANDSOME","GWAPO")
```
