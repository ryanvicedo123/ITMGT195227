## Group Output

```python
# Write code here

%matplotlib inline 

# Top 10 Countries data for graph
top10 = top10countries.sort_values(by=2019, ascending=True)

# Creating figure, size, and subplot indexes
fig, ax = plt.subplots(3,1, figsize=(12,20))
fig.set_size_inches(12, 20)

# Graph 1
ax[0].barh(y=top10["Country"],width=top10[2019]/1000, color="green")
ax[0].set_xlabel("Remittances (in Thousands)",size = 12)
ax[0].set_ylabel("Country",size = 12)
ax[0].set_title("Top 10 Sources of Cash Remittances",size = 16)

# Graph 2
ax[1].bar(x=df_continent_sorted["Country"],height=df_continent_sorted[2019]/1000, color="orange")
ax[1].set_xlabel("Continent",size = 12)
ax[1].set_ylabel("Remittances (in Thousands)",size = 12)
ax[1].set_title("Top Continents for Cash Remittances",size = 16)

# Graph 3
ax[2].bar(x=usaremit.index,height=usaremit["remittances"]/1000, color = "red")
ax[2].set_xlabel("Month",size = 12)
ax[2].set_ylabel("Remittances (in Thousands)",size = 12)
ax[2].set_title("Monthly Cash Remittances",size = 16)

# Showing graphs
plt.show()



```
