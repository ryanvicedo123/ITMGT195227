```python
maxorder=10
brewed=100
cappuccino=110
espresso=110
americano=130
qty=0
price=0
vatrate=1.013

while True:
    try:
        selection=int(input("What coffee would you like to purchase?\n1.Brewed Coffee\n2.Cappuccino\n3.Espresso\n4.Americano\n"))
        if selection<0 or selection>4:
            print("Please choose a coffee.")
            continue
    except ValueError:
        print("Please choose a coffee.")
        continue
    break
    
if selection==1:
    price=brewed
    while True:
        qty=int(input("How many Brewed Coffees would you like?"))
        if qty>maxorder or qty<0:
            print("Please choose an order below 10")
            continue
        elif qty<maxorder and qty>0:
            break

elif selection==2:
    price=cappuccino
    while True:
        qty=int(input("How many Cappucicinos would you like?"))
        if qty>maxorder or qty<0:
            print("Please choose an order below 10")
            continue
        elif qty<maxorder and qty>0:
            break

elif selection==3:
    price=espresso
    while True:
        qty=int(input("How many Expressos would you like?"))
        if qty>maxorder or qty<0:
            print("Please choose an order below 10")
            continue
        elif qty<maxorder and qty>0:
            break

elif selection==4:
    price=americano
    while True:
        qty=int(input("How many Americanos would you like?"))
        if qty>maxorder or qty<0:
            print("Please choose an order below 10")
            continue
        elif qty<maxorder and qty>0:
            break

subtotal=price*qty
total=subtotal*vatrate
print("\n\n=======================================\nYour total payment is:\n",total,"Php. Thank you and have a nice day!.")
```
